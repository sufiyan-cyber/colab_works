**Challenge**: Use <code>.groupby()</code> and <code>.count()</code> to show the number of LEGO sets released year-on-year. How do the number of sets released in 1955 compare to the number of sets released in 2019?

```python
sets_by_year = setspd.groupby('year').count()['set_num']
sets_by_year.head()
plt.plot(sets_by_year[:-2])
```

```python
thems_by_year=setspd.groupby('year').agg({'theme_id':pd.Series.nunique})

thems_by_year.rename(columns={'theme_id':'nr_themes'},inplace=True)
thems_by_year.head()
```

**Challenge**: Plot the number of themes released by year on a line chart. Only include the full calendar years (i.e., exclude 2020 and 2021).

**Challenge**: Use the <code>.groupby()</code> and <code>.agg()</code> function together to figure out the average number of parts per set. How many parts did the average LEGO set released in 1954 compared to say, 2017?

### Merging (i.e., Combining) DataFrames based on a Key


```python
sets_theme_count=pd.DataFrame({'id':set_theme_count.index,'set_count':set_theme_count.values})
sets_theme_count.head()
```

```python
merged_df=pd.merge(sets_theme_count,themes,on='id')
```

## Count Transparent and Opaque Colors (Method 1: value_counts)

### Subtask:
Count the number of transparent ('t') and opaque ('f') colors in the 'is_trans' column of the `df` DataFrame using the `.value_counts()` method.


**Reasoning**:
To count the number of transparent and opaque colors, I will select the 'is_trans' column from the `df` DataFrame and apply the `.value_counts()` method as instructed.



```python
df['is_trans'].value_counts()
```

**Reasoning**:
Now, I will count the number of transparent and opaque colors using boolean indexing as per the main task's requirement.



```python
transparent_colors = df[df['is_trans'] == 't'].shape[0]
opaque_colors = df[df['is_trans'] == 'f'].shape[0]

print(f"Number of transparent colors: {transparent_colors}")
print(f"Number of opaque colors: {opaque_colors}")
```

### Summary of Transparent and Opaque Colors

Both `value_counts()` and boolean indexing methods confirm the following distribution of colors:

- **Number of Opaque Colors (is_trans = 'f'):** 107
- **Number of Transparent Colors (is_trans = 't'):** 28
