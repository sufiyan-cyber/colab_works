```python
spread_col=df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
display(spread_col)
df.insert(1, 'Spread', spread_col)
```

```python
print("big difference is in")
themax=df['Undergraduate Major'].loc[df['Spread'].idxmax()]
display(themax)
print("smallest difference is in")
themin=df['Undergraduate Major'].loc[df['Spread'].idxmin()]
display(themin)
```

```python
sorted_values=df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
display(sorted_values.head()['Undergraduate Major'])
```

```python
sorted_spread=df.sort_values('Spread', ascending=False)
display(sorted_spread.head()['Undergraduate Major'])
```
